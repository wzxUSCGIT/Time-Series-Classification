diii)
5-fold CV to select best l:
The method is: for each l, we will have a training set. We devide it by 5, 4 for training and 1 for validation.
When do the training, use RFE to select a fixed number of features ( 5 features are selected).
Then test on the validation set to get a error rate. 
We reapeat the above procedures for 5 times and average them to get a error rate for current l.
We repeat the above procedures for 20 ls and select l with best error rate. 

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import zero_one_loss

def find_cv_score(l):
    feature_list=['mean1','mean2','mean3','mean4','mean5','mean6',
                'min1','min2','min3','min4','min5','min6',
                'max1','max2','max3','max4','max5','max6',
                 'median1','median2','median3','median4','median5','median6',
                'std1','std2','std3','std4','std5','std6',
                'q_25_1','q_25_2','q_25_3','q_25_4','q_25_5','q_25_6'
                'q_75_1','q_75_2','q_75_3','q_75_4','q_75_5','q_75_6']

    #l_list=np.arange(1,21)
    #l=3
    train_data=get_train_data(original_directory,l)
    score=np.array([0])

    skf = StratifiedKFold(n_splits=5,shuffle=True)
    X=np.array(train_data.loc[:,'max1':'std6'])
    y=np.array(train_data.loc[:,'label'])
    skf.get_n_splits(X, y)

    for train_index, test_index in skf.split(X, y):

        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        model=LogisticRegression(C=9999999) # A very large budget removes the effect of regularization
        rfe=RFE(model,5)
        #print(X_train.shape)
        rfe=rfe.fit(X_train,y_train)
        feature_index=np.array(np.where( rfe.support_== True ))[0]
        selected_feature=[]
        #for i in feature_index:selected_feature.append(feature_list[i])

        LR=LogisticRegression(C=9999999)
        LR.fit(X_train[:,feature_index],y_train)
        #print(X_train[:,feature_index].shape)
        fold_score=LR.score(X_test[:,feature_index],y_test)
        #print(fold_score)
        score=np.append(score,fold_score)
        final_score=np.mean(score[1:6])
   
    return (final_score)
```

```python
l_list=np.arange(1,21)
for l in l_list:
    
    print('L=',l,' ',find_cv_score(l))
```

From the above resut, we can see that the best L = 1 when the number of feature selected is 5. The correct method is to do the 
feature selection during the cross validation, which is described in the last text cell. The wrong method is to selecte features before we do the cross validation, ignoring the fact that feature selection is also part of the training.
